```python
import numpy as np
import tensorflow as tf
```

```python
print("tensorflow version",tf.__version__)
```

```python
path= r'book.txt'
```

```python
text = open(path, 'rb').read().decode(encoding = 'utf-8')
```

```python
print("book contains total character",len(text))
```

Prepare Text:

```python
text= text[50:]
```

Unique Character Count:

```python
vocab = sorted(set(text))
```

```python
print("Total Unique Characters:", len(vocab))
```

Character to Integer Mapping

```python
char2idx = {u:i for i, u in enumerate(vocab)}

idx2char = np.array(vocab)
```

```python
text_as_int = np.array([char2idx[c] for c in text])
```

```python
for char, _ in zip(char2idx, range(20)):
  print('{:4s}:{:3d},'.format(repr(char), char2idx[char]))
```

Text to integer Sample

```python
print(' {} ---- char-2-int ----{}'.format(repr(text[42:62]), text_as_int[40:60]))
```

### Prepare Dataset

We leverage sliding Window approch to train our model. We first chcek mac sequence length to 100 characters. (prepare for training batches)

```python
#maximum length sentence we want for a single input characters

seq_length = 100
examples_per_epoch = len(text)//(seq_length+1)

#create training example

char_dataset = tf.data.Dataset.from_tensor_slices(text_as_int)
```

```python
for i in char_dataset.take(10):
  print(idx2char[i.numpy()])
```

Prepare Batch

```python
sequences = char_dataset.batch(seq_length+1, drop_remainder=True)

for item in sequences.take(10):
  print(repr(''.join(idx2char[item.numpy()])))
  print("-"*110)
```
